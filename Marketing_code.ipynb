```python
def replace_outliers_with_median(df, column):
    # Calculate the IQR (Interquartile Range)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define the outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers and replace them with the median
    median_value = df[column].median()
    df[column] = df[column].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

    return df

df = replace_outliers_with_median(df, 'Flight Distance')
df = replace_outliers_with_median(df, 'Arrival Delay')
df = replace_outliers_with_median(df, 'Departure Delay')
```

## **Age Range Vs Passenger's Satisfaction**

```python
import matplotlib.pyplot as plt

age_category_counts = df.groupby(['Satisfaction', 'Age_Range']).size().reset_index(name='Count')

pivot_data = age_category_counts.pivot_table(index='Age_Range', columns='Satisfaction', values='Count', fill_value=0)

pivot_data_percentage = pivot_data.div(pivot_data.sum(axis=1), axis=0) * 100

pivot_data_percentage.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'salmon'])

plt.xlabel('Age_Range')
plt.ylabel('Percentage')
plt.title('Percentage of Satisfaction by Age Range')

plt.show()
```

Null Hypothesis (H0) — Age range and satisfaction of passengers are independent of each other

Alternative Hypothesis (HA) — There is an association between the age range and satisfaction of passengers

```python
from scipy.stats import chi2_contingency
import math
```

```python
grouped = df.groupby(['Age_Range', 'Satisfaction']).size().reset_index(name='num')

pivot_data = grouped.pivot_table(index='Age_Range', columns='Satisfaction', values='num', fill_value=0)

pivot_data_percentage = pivot_data.div(pivot_data.sum(axis=1), axis=0) * 100

pivot_data_percentage = pivot_data_percentage.reset_index().melt(
    id_vars='Age_Range',
    var_name='Satisfaction',
    value_name='percentage'
)

pivot_data_percentage['percentage'] = round(pivot_data_percentage['percentage'], 2)

reshaped_output = pivot_data_percentage.pivot(
    index='Age_Range',
    columns='Satisfaction',
    values='percentage'
).reset_index()

reshaped_output = reshaped_output.rename(columns={
    'Neutral or Dissatisfied': 'Dissatisfied',
    'Satisfied': 'Satisfied'
})

print(reshaped_output)
```

```python
satisfied = df[df['Satisfaction'] == 'Satisfied'].groupby('Age_Range').size().reset_index(name='satisfied')

dissatisfied = df[df['Satisfaction'] == 'Neutral or Dissatisfied'].groupby('Age_Range').size().reset_index(name='dissatisfied')

joined_df1 = pd.merge(satisfied, dissatisfied, on='Age_Range', how='outer').fillna(0)

joined_df1.index = ['Children', 'Middle Aged', 'Old', 'Youth']

joined_df1 = joined_df1[['satisfied', 'dissatisfied']]

print(joined_df1)

chi2, p, dof, expected = chi2_contingency(joined_df1)

print("\nChi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected,'\n')

if p < 0.05:
    print("The p-value is less than 0.05, indicating a significant result.")
else:
    print("The p-value is greater than or equal to 0.05, indicating no significant result.")
```

**There is sufficient evidence to support the claim that there is an association between age range and satisfaction level of passengers**
