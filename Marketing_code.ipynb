```python
def replace_outliers_with_median(df, column):
    # Calculate the IQR (Interquartile Range)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define the outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers and replace them with the median
    median_value = df[column].median()
    df[column] = df[column].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

    return df

df = replace_outliers_with_median(df, 'Flight Distance')
df = replace_outliers_with_median(df, 'Arrival Delay')
df = replace_outliers_with_median(df, 'Departure Delay')
```

Null Hypothesis (H0) â€” Age range and satisfaction of passengers are independent of each other

```python
result = df.groupby(['Flight_Dist_Range', 'Class']).size().reset_index(name='num')
result['percentage'] = result['num'] * 100 / result['num'].sum()

pivot_result = result.pivot(index='Flight_Dist_Range', columns='Class', values='percentage')

pivot_result = pivot_result[['Business', 'Economy', 'Economy Plus']]

print(pivot_result)
```

```python
thousand_1 = df[df['Flight_Dist_Range'] == '1000 - 2000'].groupby('Class').size().reset_index(name='1000 - 2000')
thousand_2 = df[df['Flight_Dist_Range'] == '2000 - 3000'].groupby('Class').size().reset_index(name='2000 - 3000')
thousand_3 = df[df['Flight_Dist_Range'] == '3000 - 4000'].groupby('Class').size().reset_index(name='3000 - 4000')
thousand_4 = df[df['Flight_Dist_Range'] == 'Below 1000'].groupby('Class').size().reset_index(name='Below 1000')

joined_df1 = pd.merge(thousand_1, thousand_2, on='Class', how='outer')
joined_df1 = pd.merge(joined_df1, thousand_3, on='Class', how='outer')
joined_df1 = pd.merge(joined_df1, thousand_4, on='Class', how='outer')

joined_df1 = joined_df1.fillna(0)

joined_df1.index = ['Business', 'Economy', 'Economy Plus']

joined_df1 = joined_df1[['1000 - 2000', '2000 - 3000', '3000 - 4000', 'Below 1000']]

print(joined_df1)

chi2, p, dof, expected = chi2_contingency(joined_df1)

print("\nChi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected, '\n')

if p < 0.05:
    print("The p-value is less than 0.05, indicating a significant result.")
else:
    print("The p-value is greater than or equal to 0.05, indicating no significant result.")
```
